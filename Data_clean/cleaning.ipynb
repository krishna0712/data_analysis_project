```python
import pyarrow
```

```python
# sometime there are repeated headers so check them and remove them
df[df['Price Each'] == 'Price Each']
```

```python
df = df[df['Order ID'] != 'Order ID']
```

```python
df['Price Each'] = df['Price Each'].astype('float64[pyarrow]')
```

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype('Int64')
```

```python
df.info()
```

```python
df.columns
```

```python
# Convert columns to standard float types
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce').astype(float)
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce').astype(float)

# Create Revenue column
df['Revenue'] = df['Quantity Ordered'] * df['Price Each']
```

```python
df.info()
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Month_Name"] = df["Order Date"].dt.strftime("%B")
```

```python
df.head(8)
```

```python
df = df.rename(columns={"Order Date": "Date"})
df = df.rename(columns = {"Quantity Ordered": "Quantity"})
```

```python
total_revenue = df["Revenue"].sum()
total_quantity = df["Quantity"].sum()
```

```python
df["Year_Month"] = df["Date"].dt.to_period("M")
```

```python
monthly_kpi = df.groupby("Year_Month").agg({
    "Revenue": "sum",
    "Quantity": "sum",
    "Order ID": "nunique"
}).reset_index()

monthly_kpi.rename(columns={
    "Order ID": "Total_Orders"
}, inplace=True)
```

```python
monthly_kpi.head(12)
```

```python
monthly_kpi["Average_Order_Value"] = (
    monthly_kpi["Revenue"] / monthly_kpi["Total_Orders"]
)

monthly_kpi["Revenue_Growth_%"] = (
    monthly_kpi["Revenue"].pct_change() * 100
)

monthly_kpi["Order_Growth_%"] = (
    monthly_kpi["Total_Orders"].pct_change() * 100
)
```

```python
monthly_kpi.head()
```

```python
monthly_kpi.fillna(0, inplace=True)
```

```python
monthly_kpi.head(13)
```

```python
product_kpi = df.groupby("Product").agg({
    "Revenue": "sum",
    "Quantity": "sum",
    "Order ID": "nunique"
}).reset_index()

product_kpi.rename(columns={
    "Order ID": "Total_Orders"
}, inplace=True)

product_kpi = product_kpi.sort_values("Revenue", ascending=False)
```

```python
product_kpi.head()
```

```python
total_revenue = product_kpi["Revenue"].sum()

product_kpi["Revenue_%"] = (
    product_kpi["Revenue"] / total_revenue * 100
)
```

```python
product_kpi.head()
```

```python
concentration_risk = product_kpi["Revenue_%"].max() > 40
```

```python
monthly_kpi["Decline"] = monthly_kpi["Revenue"].diff() < 0

two_month_decline = (
    monthly_kpi["Decline"].rolling(2).sum().max() == 2
)
```

```python
growth_slowing = (
    monthly_kpi["Revenue_Growth_%"].diff() < 0
).sum() >= 2
```

```python
monthly_kpi.head(12)
```

```python
print("Concentration Risk:", concentration_risk)
```
